```python
import csv
import matplotlib.pyplot as plt

data_csv='population-estimates.csv'
def read_data(csv): 
    # Reading Data from csv
    with open(data_csv,'r') as f:
        csv_reader = csv.DictReader(f)
        population_data_list=list(csv_reader)
    return population_data_list

def get_Indian_population():
    data=read_data(csv)
    indian_population_over_years={}
    for row in data:
        if row['Region']=='India':
            indian_population_over_years[row['Year']]=row['Population']
    return indian_population_over_years

def plot_Bar_chart():
    data=get_Indian_population()
    years=list(data.keys())
    population=list(data.values())
    fig = plt.figure(figsize = (16,9))
    # creating the bar plot
    plt.gcf().autofmt_xdate()
    plt.bar(years,population[0:10000], color ='maroon')
    plt.xlabel("years")
    plt.title("Indias population over years")

def execution_population_data_india():
    plot_Bar_chart()

execution_population_data_india()
```

```python
Asean_countries=['Brunei Darussalam','Cambodia','Indonesia',"Lao People's Democratic Republic",'Malaysia','Myanmar','Philippines','Singapore', 'Thailand','Viet Nam']

def get_Asean_population_2014():
    data=read_data(csv)
    Asean_population_2014={}
    for row in data:
        if row['Region'] in Asean_countries and row['Year']=='2014':
            Asean_population_2014[row['Region']]=int(float(row['Population']))
    Asean_population_2014=dict(sorted(Asean_population_2014.items(), key=lambda x:x[1]))
    return Asean_population_2014

def plot_Bar_chart():
    data=get_Asean_population_2014()
    countries=list(data.keys())
    population=list(data.values())
    fig = plt.figure(figsize = (16,9))
    # creating the bar plot
    plt.gcf().autofmt_xdate()
    plt.bar(countries,population[0:5000], color ='maroon')
    plt.xlabel("countries")
    plt.title("Asean countries population over years")

def execution_Asean_countries_population_2014():
    plot_Bar_chart()

execution_Asean_countries_population_2014()
```
